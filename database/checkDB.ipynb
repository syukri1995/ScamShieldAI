```python
import pandas as pd
```

```python
import sqlite3

# Load scamshield.db
conn = sqlite3.connect("scamshield.db")
cursor = conn.cursor()

# Optional quick check: list tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Loaded scamshield.db")
print("Tables:", [t[0] for t in tables])
```

```python
from typing import Any

def get_table_info(scan_history: str) -> Any:
    """Fetch all records from a specified table."""
    cursor.execute(f"SELECT * FROM {scan_history}")
    return cursor.fetchall()


df = pd.DataFrame(get_table_info("scan_history"))
print(df.head())
```
